```python
import pandas as pd
import json
from pathlib import Path
import numpy as np

parquet_dir = Path("data/battles_raw_batches")
parquet_files = sorted(parquet_dir.glob("*.parquet"))
if not parquet_files:
    print("Nenhum arquivo Parquet encontrado.")
else:
    df = pd.read_parquet(parquet_files[0])
    # Seleciona uma batalha aleat√≥ria
    random_battle = df.sample(n=1).iloc[0].to_dict()

    def make_serializable(obj):
        if isinstance(obj, np.ndarray):
            return make_serializable(obj.tolist())
        if isinstance(obj, (np.generic,)):
            return obj.item()
        if isinstance(obj, pd.Timestamp):
            return obj.isoformat()
        if isinstance(obj, pd._libs.tslibs.nattype.NaTType):
            return None
        if isinstance(obj, dict):
            return {k: make_serializable(v) for k, v in obj.items()}
        if isinstance(obj, list):
            return [make_serializable(i) for i in obj]
        return obj

    random_battle = make_serializable(random_battle)
    print(json.dumps(random_battle, indent=2, ensure_ascii=False))
```
